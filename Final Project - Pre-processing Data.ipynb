```python
import pandas as pd
import numpy as np
import os
```

#### Data Source: Kaggle dataset

https://www.kaggle.com/borismarjanovic/price-volume-data-for-all-us-stocks-etfs/data

```python
# create the tickers list
tickers=[]
for filename in os.listdir('./Data/Stocks'):
    file_path = os.path.join('./Data/Stocks/', filename)
    ticker = filename.split('.')[0]
    tickers.append(ticker)
```

```python
# define function to generate df given list of tickers and date range
def stocks_data(symbols, dates):
    df = pd.DataFrame(index=dates)
    for symbol in symbols:
        try:
            print ("Working on ticker:",symbol)
            df_temp = pd.read_csv("./Data/Stocks/{}.us.txt".format(symbol), index_col='Date',
                    parse_dates=True, usecols=['Date', 'Close'], na_values=['nan'])
            df_temp = df_temp.rename(columns={'Close': symbol})
            df = df.join(df_temp)
        except:
            continue
    return df
```

```python
# get the stock data for the date range
dates = pd.date_range('2012-01-01','2016-12-31',freq='B')
df_stocks = stocks_data(tickers, dates)
```

```python
df_stocks
```

```python
# remove NaN's
df_stocks = df_stocks.fillna(method='pad')
df_stocks = df_stocks.fillna(method='backfill')
```

```python
# verify NaN's
df_stocks.isnull().values.sum()
```

```python
# remove NaN's - all
df_stocks_clean=df_stocks.dropna(axis=1,how='all')
```

```python
# verify NaN's on cleaned dataset
df_stocks_clean.isnull().sum().sum()
```

```python
# write to a csv file
df_stocks_clean.to_csv('stocks_clean.csv')
```
